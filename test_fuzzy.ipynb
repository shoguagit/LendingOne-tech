```python
from fuzzywuzzy import process
import pandas as pd


df_transactions = pd.read_csv("transactions.csv")
df_customers = pd.read_csv("customers.csv")


def valida_dato(nombre_cliente, lista_clientes, porcentaje_acierto=100):
    result = process.extractOne(nombre_cliente, lista_clientes)
    return result[0] if result and result[1] >= porcentaje_acierto else None

customer_names = df_customers['customer_name'].unique()

df_transactions['matched_name'] = [
    valida_dato(nombre, customer_names) for nombre in df_transactions['customer_name']
]

df_consolidado = pd.merge(
    df_transactions, 
    df_customers, 
    left_on='matched_name', 
    right_on='customer_name', 
    how='left', 
    suffixes=('_transaction', '_customer')
)

df_consolidado
```
